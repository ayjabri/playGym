## The eight queens puzzle:

The eight queens puzzle, or the eight queens problem, asks how to place eight queens on a chessboard without attacking each other. If you never played chess before, a queen can move in any direction (horizontally, vertically and diagonally) any number of places. In the next figure, you can see two queens with their attack patterns:


If we represent the grid in a set of parallal diagonal lines with intercepts in range(N), then a valid solution is one where each line contains one queen only.
for example: the first diagonal line's equation is y = x + 0 => y - x = 0

Second line is y - x =1. Third y - x = 2...etc.


Applying the same concept we conclude that a valid solution is one that the result of :
y - x is in (0,1,2,3,4,5,6,7)
and the result of 
y + x is in (0,1,2,3,4,5,6,7). for the other direction



To solve this problem we need first to represent the locations. A common approach is to use column location such as:

(0,2,4,5,7,6,3,1)

```python
N = 8
y = range(N)
solution = []
for state in permutations(y):
    if len(set(state[i]+i for i in y))==len(set(state[i]-i for i in y))==N:
        solution.append(state)
```

```python
import numpy as np
s = np.zeros((len(solution),N,N),dtype=str)
for n in range(len(solution)):
    for i in y:
        s[n][i,solution[n][i]]='X'
```
